# format() 함수
 - 문자열 형태가 있는 함수를 사용함.
 - "{자료형}".format(인수)
 - 자료형을 바로 지정해 주지 않고 변수가 할당되는 장점!

```python
print("I'm {0}yearsold.".format(20))
```

```python
age=40; name="zimean"
print("I'm {0}yearsold.".format(age))
print("My name is {0} and {1} years old.".format(name,age))
print("Product: {0},Price per unit :{1:.2f}.".format("Apple",5.243)) 
```

 - {1:.2f} .2f: 소수점 둘째 자리까지 출력하라.

# 패딩 : 여유공간을 지정하여 글자 배열을 맞추고 소수점 자릿수를 맞추는 기능

## % 서식의 패딩 

```python
print("%10d" %12)  # 10자리 확보하고 우측 정렬 
print("%-10d" %12) # 좌측 정렬 9번째칸 부터 12 출력

print("%10.3f" %5.94343)  # 10자리 확보하고 소수점 셋째자리까지 출력
```

## format()함수의 패딩 

```python
print("{0:>10s}".format("Apples")) #10자리 확보하고 우측정렬로 Apple
print("{0:<10s}".format("Apples")) #10자리 확보하고 죄측정렬로 Apple

print("{1:10.5f}.".format("Apple",5.243))
```

```python
"{1:>10.5f}.".format("Apple",5.243)
```

```python
"{1:<10.5f}.".format("Apple",5.243)
```

## 네이밍

```python

print("Product: %(name)5s, Price per unit:%(price)5.5f."%{"name":"Apple","price":5.243})
print("Product : {name:>5s},Price per unit:{price:>5.5f}.".format(name="Apple",price=5.243))
```
