# (1) deque 모듈
- 스택과 큐 모두 지원하는 모듈
- 리스트와 비슷한 형식으로 데이터를 저장해야함 append()
- pop()하면 스택처럼 나중에 넣은 값부터 하나씩 추출할 수 있음
- pop(0)은 안됨.-> appendleft() 새로운 값을 왼쪽부터 입력되게 하여 먼저 들어간 값부터 출력될 수 있도록 할 수 있다.

```python
from collections import deque
```

```python
deque_list=deque()
print(deque_list)
for i in range(5):
    deque_list.append(i)
print(deque_list)
```

```python
deque_list.pop()
```

```python
deque_list.pop()
```

```python
deque_list.pop()
```

```python
deque_list
```

```python
for i in range(5):
    deque_list.appendleft(i)
print(deque_list)
```

```python
deque_list.pop()
```

```python
deque_list.popleft()
```

# (1)-1 rotate() 함수 


```python
from collections import deque

deque_list=deque()
for i in range(5):
    deque_list.appendleft(i)
print(deque_list)

#rotate()
deque_list.rotate(2)
print(deque_list)
```

- reversed()
- extend()
- extendleft()

```python
print(deque(reversed(deque_list)))
print(deque_list)
```

```python
deque_list.extend([5,6,7])
print(deque_list)
```

```python
deque_list.extendleft([5,6,7])
print(deque_list)
```

# (3) defaultdict 모듈
- 딕셔너리의 변수를 생성할 때 키에 기본 값을 지정하는 방법
 * 실제로 키를 생성하지 않고 해당 키의 값을 호출하려할때 오류가 발생한다.

```python
d= dict()
print(d["first"])
```

```python
from collections import defaultdict

d=defaultdict(lambda:0)  
print(d["first"])
```

- lambda() : return 0과 같다. 어떤 키가 들어오더라도 처음 값은 전부 0으로 설정한다는 뜻.

```python
from collections import defaultdict

s=[('yellow',4),('blue',2),('yellow',3),('blue',4),('red',1)]
d=defaultdict(list)
print(d)
for k,v in s:
    
    d[k].append(v)
#    print(d[2].append(v))
print(d.items())
```

### 딕셔너리 형태로 Counter 객체를 생성하는 방법
- elements() : 각 요소의 갯수만큼 리스트형의 결과 출력

```python
from collections import Counter

c=Counter({'red':4,'blue':2})
print(c)
```

```python
print(list(c.elements()))
```

### 키워드 형태의 매개변수를 사용하여 Counter 를 생성하는 방법
- 매개변수의 이름을 키
( 로 , 실제 값을 값 value) 으로 하여 Counter 를 생성

```python
from collections import Counter

c=Counter(cats=4,dogs=8)
print(c)
```

```python
print(list(c.elements()))
```

### Counter의 기본 사칙연산

```python
from collections import Counter

c=Counter(a=4,b=2,c=0,d=-2)
d=Counter(a=1,b=2,c=2,d=4)
c.subtract(d)

c
```

```python
text='''People infected by the novel coronavirus tend to develop symptoms about five days after exposure, and almost 
always within two weeks, according to a study released Monday.

That incubation period is consistent with previous estimates from public health officials, 
and the findings suggest that 14 days of quarantine are appropriate for people potentially 
exposed to the coronavirus.
'''.lower().split()
```

```python
from collections import OrderedDict
for i,v in OrderedDict(sorted(word_count.items(), key=lambda t:t[1],reverse=True)).items():
    print(i,v)
```
